##브로드캐스트 규칙

1. 기본원칙: 모양이 같아야함

기본적으로 요소별로 산술연산이 이루어지기에 짝이 맞아야함

<br>

2. 예외조항: 브로드캐스팅

넘파이는 연산하려는 두 배열의 모양이 달라도, 두 가지 조건 중 하나를 만족하면 큰 배열의 모양에 맞게 작은 배열의 모양을 확장시켜 연산을 수행함

**[규칙1] 두 배열의 차원 수가 다르면, 차원이 낮은 배열의 앞(왼쪽)을 1로 채움**

**[규칙2] 각 차원에서 두 배열의 크기가 같거나, 어느 한쪽의 크기가 1이어야함**

```python
#브로드캐스트
A = np.array([[1,2], [3,4]]) #2차원배열
B = np.array([10,20]) #1차원배열

print(A*B)

print(A.shape)
print(B.shape)
```

```python
#브로드캐스트: 적용되지 않는 경우
A = np.array([[1,2], [3,4], [5,6]])
B = np.array([[10,20], [30,40]])

A*B
```

```python
#브로드캐스트: 적용되지 않는 경우 => 값이 다른데 1인쪽이 없음 (3,2) (1,2)면 가능
A = np.array([[1,2], [3,4], [5,6]])
B = np.array([[10,20], [30,40]])

print(A.shape)
print(B.shape)
```

```python
#원소 접근
X = np.array([[51,55], [14,19], [0,4]])
print(X)

print(f"X[0]: {X[0]}") #0행
print(f"X[0][1]: {X[0][1]}") #(0,1) 위치의 원소
```

```python
#원소 접근: for문

for row in X:
  print(row)
```

```python
#flatten: 평탄화
X = np.array([[51,55], [14,19], [0,4]])
X = X.flatten()

print(X)
```

```python
#인덱스가 0,2,4인 원소 얻기 (앞에서 1차원 배열로 바꿨으니까)
X[np.array([0,2,4])]
```

```python
#특정 조건만 만족하는 원소 얻기
print(X > 15) #조건에 대한 각 원소의 T/F 나옴
print(X[X>15])
```

## 맷플롯립 (matplotlib)

```python
#1. 단순 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt

#데이터 준비
x = np.arange(0,6, 0.1) #0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

#그래프 그리기
plt.plot(x,y)
plt.show()
```

```python
from matplotlib.lines import lineStyles
#2. pylot의 기능

#데이터 준비
x = np.arange(0,6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") #cos함수는 점선으로 그리기
plt.xlabel("x") #x축 이름
plt.ylabel("y") #y축 이름
plt.title("sin&cos") #제목

plt.legend()
plt.show()
```
